```python
import pandas as pd
import numpy as np
import sklearn as skl
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import IsolationForest
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt

print("Environment is ready!")  
```

```python
# טעינת קובץ דאטה
filePath = '../data/raw/Wednesday-workingHours.pcap_ISCX.csv'

print(f"Loading data from {filePath}...")
dataFile = pd.read_csv(filePath)

dataFile.columns = dataFile.columns.str.strip()

print("Data loaded successfully!")
print(f"Data shape: {dataFile.shape}")
```

```python
dataFile.head()
print(dataFile['Label'].value_counts())
```

```python
print("Missing values:", dataFile.isnull().sum().sum())
print("Infinite values:", np.isinf(dataFile.select_dtypes(include=np.number)).sum().sum())

original_rows = dataFile.shape[0]

dataFile.replace([np.inf, -np.inf], np.nan, inplace=True)

dataFile.dropna(inplace=True)
print(f"Original rows: {original_rows}")
print(f"Current rows:  {dataFile.shape[0]}")
print(f"Dropped rows:  {original_rows - dataFile.shape[0]}")

print("Remaining NaNs:", dataFile.isnull().sum().sum())
```
